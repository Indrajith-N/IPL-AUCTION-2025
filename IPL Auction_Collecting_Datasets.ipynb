### IMPORTING PACKAGES

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.iplt20.com/auction"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

## CSK AUCTION DATA

```python
section1 = soup.find('section',{'class':"auction-tab-section auction-tabbed_panel ih-vbg-n ap-latest-news ap-no-bg-result col-100 floatLft relative"})
div=section1.find('div',{'id':"autab3-2022"})
section2=div.find('section' ,{'class':"ih-points-table-sec position-relative" ,'id':"CSK-inside"})
table=section2.find('table',{'id':"t3-chennai-super-kings"})

headers = []
for header in table.find_all('th'):
    headers.append(header.text.strip())
cells=[]
for row in table.find_all('tr')[1:]:  # Skip the header row
    for cell in row.find_all('td'):
        cells.append(cell.text.strip())
data=[]
for i in range(0,len(cells),6):
    data.append(cells[i:i+6])
df1=pd.DataFrame(data,columns=headers)

```

## Collecting Images from Websites


```python
url = "https://www.iplt20.com/auction"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
section=soup.find('section',{'class':'auction-tab-section auction-tabbed_panel ih-vbg-n ap-latest-news ap-no-bg-result col-100 floatLft relative'})
div=section.find('div',{'class':'auc-round-logo'})
image=[]
for a in div.find_all('a'):
    for img in a.find_all('img'):
        image.append(img['src'])
filename=['CSKroundbig.png',
 'DCroundbig.png',
 'GTroundbig.png',
 'KKRroundbig.png',
 'LSGroundbig.png',
 'MIroundbig.png',
 'PBKSroundbig.png',
 'RRroundbig.png',
 'RCBroundbig.png',
 'SRHroundbig.png']
```

```python
try:
    for i in range(len(image)):
        response=requests.get(image[i])
        with open(filename[i],'wb')as file:
            file.write(response.content)
except:
    print("FILE ALREADY EXISTS")
    
```

```python
IMG="https://documents.iplt20.com//ipl/assets/images/ipl-logo-new-old.png"
FILENAME='ipl-logo-new-old.png'
try:
    response=requests.get(IMG)
    with open(FILENAME,'wb')as file:
        file.write(response.content)
except:
    print("FILE ALREADY EXISTS")
```
